```python
# Removing columns with many null values
hotel_bookings.drop(['agent','company'],axis=1,inplace=True)
```

```python
# Removing some outliers
hotel_bookings=hotel_bookings[hotel_bookings['adr']<5000]
```

```python
# Cancelation %
cancelled_perc = hotel_bookings['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plot = sns.countplot(x='is_canceled', data = hotel_bookings)

# Changing xticks
xtick_loc = plot.get_xticks()
xtick_labels = plot.get_xticklabels()

# Set the x-axis ticks
plot.set_xticks(ticks = xtick_loc, labels = ["Not Canceled", "Canceled"])
```

```python
plt.figure(figsize=(8,4))
plot = sns.countplot(x="hotel",hue="is_canceled", data = hotel_bookings, palette = "Blues")
plt.title("Reservation Status in different hotels")
plt.xlabel("Hotel")
plt.ylabel("Number of reservations")
plt.legend(["not canceled","canceled"])
plt.show()
```

We can see the fact that city hotels have more bookings, but more cancelation rate as well

```python
resort_hotel_pricing = hotel_bookings[hotel_bookings["hotel"]=="Resort Hotel"].groupby("reservation_status_date")[['adr']].mean()
city_hotel_pricing = hotel_bookings[hotel_bookings["hotel"]=="City Hotel"].groupby("reservation_status_date")[['adr']].mean()
```

```python
plt.figure(figsize=(20,10))
plt.title("Avg Daily Rate for resort and city hotel",fontsize = 20)
plt.plot(resort_hotel_pricing.index,resort_hotel_pricing["adr"])
plt.plot(city_hotel_pricing.index,city_hotel_pricing["adr"])
```

The above graphs clears the fact that adr i.e. average daily rate for resort is more distorted than city hotels.

```python
hotel_bookings["month"]=hotel_bookings["reservation_status_date"].dt.month
plt.figure(figsize=(20,10))
plt.title("Month vs booking-cancelation",fontsize=20)
sns.countplot(x='month', hue = 'is_canceled',data=hotel_bookings)
```

We can see that the most cancelation takes place in January, but least in August and September.

```python
plt.figure(figsize=(20,10))
sns.barplot(x="month",y="adr",data=hotel_bookings[hotel_bookings["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
plt.title("ADR per month", fontsize=20)
```

As we show cancelation are more in January, so is the ADR and similarly lowest adr are in August and September so is the cancelation rate

```python
top_10_countries = hotel_bookings[hotel_bookings["is_canceled"]==1]["country"].value_counts()[:10]
plt.figure(figsize=(10,5))
plt.title("Top 10 counties with cancelation")
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)
```

Pie chart cearly showcases the most number of cancelations takes place in Portugal by far.

```python
# Checkng adr's for cancelation and non-canceled bookings
is_canceled_adr = hotel_bookings[(hotel_bookings["is_canceled"]==1) & (hotel_bookings["reservation_status_date"]>"2016") & (hotel_bookings["reservation_status_date"]<"2017-09")].groupby("reservation_status_date")[["adr"]].mean()
not_canceled_adr = hotel_bookings[(hotel_bookings["is_canceled"]==0) & (hotel_bookings["reservation_status_date"]>"2016") & (hotel_bookings["reservation_status_date"]<"2017-09")].groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(20,10))
plt.title("ADR of canceled and non-canceled bookings", fontsize=20)
plt.plot(is_canceled_adr.index,is_canceled_adr["adr"],label="Is Canceled")
plt.plot(not_canceled_adr.index,not_canceled_adr["adr"],label="Not Canceled")
plt.legend()
```

This line plot clearly demonstrates more the adr, more the cancelation and above EDA also proves City otel and Portugal are major contributors as well. Thus, by monitoring adr most of the cancelation can be contained.
